What does it mean from the financial point of view for the stock (you can get bonus
partial points)?

```python
#the list of companies that have same close and adj close
same_close_adj_close = []
```

```python

for company_ticker, df_data  in data_frames.items():
    number_of_row = len(df_data)

    #the rows that Close is equal to Adj Close

    equal_close_adjClose =  df_data[df_data['Close'] == df_data['Adj Close']]

    display(equal_close_adjClose)

    if len(equal_close_adjClose) == number_of_row:
        same_close_adj_close.append(company_ticker)
        print(company_ticker)
```

What is the highest and lowest price (Adj Close) each company recorded?

```python
for company_ticker, df_data  in data_frames.items():


    print('Company tick : ', company_ticker, ' Lowest price : ', df_data['Adj Close'].min(), ' Highest price : ', df_data['Adj Close'].max())
```

(1pt task) Calculate logarithmic returns from Adj Close. For each company report on
its, min, max, mean, median of the return distribution.

Reference:   https://stackoverflow.com/questions/31287552/logarithmic-returns-in-pandas-dataframe

```python
for company_ticker, df_data  in data_frames.items():
    df_data['log_ret'] = np.log(df_data['Adj Close']) - np.log(df_data['Adj Close'].shift(1))
    
    print('Company tick : ', company_ticker, ' min return : ', df_data['log_ret'].min(), 
          ' max price : ', df_data['log_ret'].max(), ' mean return : ', df_data['log_ret'].mean(numeric_only=True), 
          ' median return : ', df_data['log_ret'].median(numeric_only=True))
```

7. Plot the log-returns of the companies (ideally in the same plot)

Reference: https://stackoverflow.com/questions/45574099/plot-different-columns-of-different-dataframe-in-the-same-plot-with-pandas

```python
fig, ax = plt.subplots()
ax2 = ax.twinx()

# plt.fig(figsize=(19,2))

for company_ticker, df_data  in data_frames.items():

    df_data.plot(x = 'Date', y = ['log_ret'], ax=ax)
fig.set_figwidth(15)
plt.legend(files, title='Companies')

plt.show()
```

Show the log-return distribution of the companies’ returns (ideally in the same plot

```python
fig, ax = plt.subplots()
ax2 = ax.twinx()

# plt.fig(figsize=(19,2))

for company_ticker, df_data  in data_frames.items():

    df_data.plot(kind = 'hist', y = ['log_ret'], ax=ax)
fig.set_figwidth(15)
plt.legend(files, title='Companies')

plt.show()
```
